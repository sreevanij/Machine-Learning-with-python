```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
rnd = np.random.RandomState(1)
rnd
x = 10*rnd.rand(100)
y = 2*x+rnd.rand(100)-5
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x[:,np.newaxis],y)
```

```python
predictx = np.linspace(0,10,1000)
predicty = model.predict(predictx[:,np.newaxis])
plt.scatter(x,y)
plt.scatter(predictx,predicty)
```

```python
model.score(x[:,np.newaxis],y) #Rsquared
```
